```python
ReprTandAll = np.zeros((50, 2, 100))



ReprTandAll[0] = ReprRand

ReprTandAll[0]

for i in range (50):
        ReprTandAll[i] = np.random.randn(2, 100)+(i/50)
```

```python
for i in range(50):
    
    plt.scatter(ReprTandAll[i][0],ReprTandAll[i][1])
```
