# Do it for the Binomial distribution

```python
##your code here...
distributions["binomial"] = {"dist":[],"mean":{}}
 
for i in range(1,100):
    distributions["binomial"]["dist"].append(np.random.binomial(n=100,p=0.1,size=i*20))
    distributions["binomial"]["mean"][i*20]=(mean(np.random.binomial(n=100,p=0.1,size=i*20)))
    #plt.hist(distributions["binomial"]["dist"][i-1],alpha = 0.5)
    #plt.xlabel('sample mean', fontsize=18)
    #plt.ylabel('sample size', fontsize=18)
    #plt.title('Figure 3 : Binomial', fontsize=18)
#plt.figure(figsize=(20,50))
```

# Do it for Lapalace distribution

```python
##your code here...
distributions["lap"] = {"dist":[],"mean":{}}
for i in range(1,100):
    distributions["lap"]["dist"].append(np.random.laplace(loc=10, scale=1.0, size=i*20))
    distributions["lap"]["mean"][i*20]=(mean(np.random.laplace(loc=10, scale=1.0, size=i*20)))
    
    #plt.hist(distributions["lap"]["dist"][i-1],alpha = 0.5)
    #plt.xlabel('sample mean', fontsize=18)
    #plt.ylabel('sample size', fontsize=18)
    #plt.title('Figure 4 : Laplace', fontsize=18)
```

```python
distributions["chaisq"] = {"dist":[],"mean":{}}
for i in range(1,100):
    distributions["chaisq"]["dist"].append(np.random.chisquare(df=10,size=i*20))
    distributions["chaisq"]["mean"][i*20]=(mean(np.random.chisquare(df=10,size=i*20)))
    #plt.hist(distributions["chaisq"]["dist"][i-1],alpha = 0.5)
    
    #plt.xlabel('sample mean', fontsize=18)
    #plt.ylabel('sample size', fontsize=18)
    #plt.title('Figure 5 : Chi Squared', fontsize=18)
```

# Scatter plot all sample means & sample size

```python
fig = pl.figure(figsize=(15,10))
i = 1
dis_titles = ['normal', 'binomial', 'pois', "lap", "chaisq"]
for dist in dis_titles:
    #print(dist)
    ax_mu_n = fig.add_subplot(2,3,i)
    
    for n in distributions[dist]["mean"]:
        #print(shape(n))
        ax_mu_n.plot(n, distributions[dist]["mean"][n], 'o')
        ax_mu_n.set_xlabel('sample size', fontsize=18)
        ax_mu_n.set_ylabel('sample mean', fontsize=18)
        ax_mu_n.set_title(dist, fontsize=18)
        ax_mu_n.plot([0, 2000], [mean_number/10, mean_number/10], 'k')
    i += 1
plt.tight_layout()
```

*** Central Limit Theorem***

When the sample size is growing, the sample mean will be closer to the mean that we set in the random function.
In each distributions, we all can see the larger sample size, the closer the mean to 10 which is the mean I set in the beginning.
